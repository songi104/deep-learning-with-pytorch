### 3.1.1 K-nearest neighbor

```python
import numpy as np
import matplotlib as plt
import pandas as pd
from sklearn import metrics
```

```python
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
dataset = pd.read_csv("data/iris.data", names=names)
full_data = pd.read_csv("data/iris.data")
```

```python
dataset
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 4].values
X, y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) # 8/2의 비율로 train, test 나눔
```

```python
from sklearn.preprocessing import StandardScaler
s = StandardScaler()
X_train = s.fit_transform(X_train) # scailing 처리
X_test = s.fit_transform(X_test)
```

```python
np.std(X_train, axis=0)
```

```python
# 모델 생성 및 훈련
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=50)
knn.fit(X_train, y_train)
```

```python
# 정확도 출력
from sklearn.metrics import accuracy_score
y_pred = knn.predict(X_test)
print(f"정확도: {accuracy_score(y_test,y_pred)}")
```

```python
k = 20
accs = [0]
for i in range(1, k+1):
    classifier = KNeighborsClassifier(n_neighbors=i).fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    acc = metrics.accuracy_score(y_test, y_pred)
    accs.append(acc)
```

```python
accs
```

```python
npAccs = np.array(accs, dtype=float)  # np.array()를 사용해 변환
maxAcc = np.max(npAccs)  # 최대 정확도 찾기
idx = np.argmax(npAccs)  # 최대 정확도의 인덱스 찾기
print(f"최대 정확도는 {maxAcc}이고 이 때 k는 {idx}입니다")
```

### 3.1.2 서포트 벡터 머신
붓꽃 분류 문제를 서포트 벡터 머신으로 풀어봅시다

```python
from sklearn import svm, metrics, datasets, model_selection
import tensorflow as tf
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
```

```python
iris = datasets.load_iris()
X_train, X_test, y_train, y_test = model_selection.train_test_split(iris.data,
                                                                    iris.target,
                                                                    test_size = 0.6,
                                                                    random_state = 42)
```

```python
svm = svm.SVC(kernel='linear', C=1.0, gamma = 0.5)
svm.fit(X_train, y_train)
```

```python
predictions = svm.predict(X_test)
score = metrics.accuracy_score(y_test, predictions)
```

```python
print(f"정확도: {score:f}")
```

### 3.1.3 결정 트리
불확실성을 줄이도록 트리모양을 학습한다.
불확실성을 계산하는 방법에는 엔트로피, 지니 계수가 있다.

```python
df = pd.read_csv("../ch03/data/titanic/train.csv", index_col='PassengerId')
print(df.head())
```

```python
df = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Survived']]
df['Sex'] = df['Sex'].map({'male':0, 'female':1})
df = df.dropna() # 값이 없는 데이터 삭제
```

```python
# df.head()
X = df.drop('Survived', axis=1)
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predict)
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
pd.DataFrame(
    confusion_matrix(y_test, y_predict),
    columns=['Predicted Not Survived', 'Predicted Survival'],
    index = ['True Not Survival', 'True Survival']
)
```
